```python
image_dir=Path('../input/a-large-scale-fish-dataset/Fish_Dataset/Fish_Dataset')
```

```python
filepaths=list(image_dir.glob(r'**/*.png'))
filepaths
```

```python
labels=list(map(lambda x:os.path.split(os.path.split(x)[0])[1],filepaths))
labels
```

```python
filepaths=pd.Series(filepaths,name='filepath').astype(str)
labels=pd.Series(labels,name='label')
images = pd.concat((filepaths,labels),axis=1)
images
```

```python
images['label']=images['label'].apply(lambda x:np.NaN if x[-2:]=='GT' else x)
```

```python
images=images.dropna()
images
```

```python
images.sample(200*9)['label'].value_counts()
```

```python
samples=[]
for category in images['label'].unique():
    category_slice=images.query("label==@category")
    samples.append(category_slice.sample(200,random_state=1))
images=pd.concat(samples,axis=0)
images
```

```python
images['label'].value_counts()
```

```python
train_df,test_df=train_test_split(images,train_size=7,random_state=1)
```

```python
trainGenerator=tf.keras.preprocessing.image.ImageDataGenerator(preprocessing_function=tf.keras.applications.mobilenet_v2.preprocess_input,
                                                              validation_split=0.2)
testGenerator=tf.keras.preprocessing.image.ImageDataGenerator(preprocessing_function=tf.keras.applications.mobilenet_v2.preprocess_input,
                                                             )
```

```python
train_images=trainGenerator.flow_from_dataframe(dataframe=train_df,x_col='filepath',y_col='label',target_size=(224,224),color_model='rgb',class_model='categorical',batch_size=32,shuffle=True,seed=42,subset='training')
val_images=trainGenerator.flow_from_dataframe(dataframe=train_df,x_col='filepath',y_col='label',target_size=(224,224),color_model='rgb',class_model='categorical',batch_size=32,shuffle=True,seed=42,subset='validation')
test_images=trainGenerator.flow_from_dataframe(dataframe=train_df,x_col='filepath',y_col='label',target_size=(224,224),color_model='rgb',class_model='categorical',batch_size=32,shuffle=False)
```

```python
pretrainedmodel=tf.keras.applications.MobileNetV2(input_shape=(224,224,3),include_top=False,weights='imagenet',pooling='avg')
pretrainedmodel.trainable=False
```

```python
inputs=pretrainedmodel.input
x=tf.keras.layers.Dense(128,activation='relu')(pretrainedmodel.output)
x=tf.keras.layers.Dense(128,activation='relu')(x)

outputs=tf.keras.layers.Dense(5,activation='softmax')(x)
model=tf.keras.Model(inputs=inputs,outputs=outputs)

model.compile(
optimizer='adam',
loss='categorical_crossentropy',
metrics=['accuracy'])

history=model.fit(
train_images,validation_data=val_images,epochs=100,callbacks=[
    tf.keras.callbacks.EarlyStopping(monitor='val_loss',patience=3,restore_best_weights=True)
])
```

```python
results=model.evaluate(test_images,verbose=0)
print("Test Loss:{:.5f}".format(results[0]))
print("Test Accuracy:{:.2f}%".format(results[1]*100))
```
