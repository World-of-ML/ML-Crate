```python
file = open("sherlockholmes.txt", "r", encoding = "utf8")
old_lines = []
for i in file:
    old_lines.append(i)

lines = []
z = len(old_lines)
for i in range(0,len(old_lines)):
  if i>57 and i<z-373:
    lines.append(old_lines[i])

data = ""
for i in lines:
  data = ' '. join(lines) 

data = data.replace('\n', '').replace('\r', '').replace('\ufeff', '').replace('“','').replace('”','').replace('I.','').replace('_','')

data = data.split()
data = ' '.join(data)
print("Total Characters in Dataset = " + str(len(data)))
print(data[0:500])

text_file = open("data.txt", "w")
text_file.write(data)
text_file.close()
```

```python
happy_wp = HappyWordPrediction()
```

```python
from happytransformer import WPTrainArgs
args = WPTrainArgs(num_train_epochs=3)
happy_wp.train("data.txt", args=args)
```

```python
text = "Sherlock Holmes is going to be the best [MASK]."

result = happy_wp.predict_mask(text, top_k=2)
print(result[0]) 
print(result[0].token) 
print(result[0].score)  
```
