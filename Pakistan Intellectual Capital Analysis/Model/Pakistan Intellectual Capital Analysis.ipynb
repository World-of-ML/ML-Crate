# Visualisation of Data

```python
base_color = sns.color_palette()[0]
order= df['Department'].value_counts().index
sns.countplot(data=df,y='Department',order=order, color=base_color)
plt.title('Teachers BY Department',size=20);
```

```python
plt.figure(figsize=[20,30])
order= df['Terminal Degree'].value_counts().index
sns.countplot(data=df,y='Terminal Degree',order=order)
plt.title('Teachers BY Terminal Degree',size=20);
```

```python
plt.figure(figsize=[20,30])
order= df['Designation'].value_counts().index
sns.countplot(data=df,y='Designation',order=order)
plt.title('Teachers BY Designation',size=20);
```

```python
plt.figure(figsize=[20,30])
order= df['Country'].value_counts().index
sns.countplot(data=df,y='Country',order=order)
plt.title('Teachers BY Country',size=20);
```

```python
plt.figure(figsize=[20,30])
order= df['Year'].value_counts().index
sns.countplot(data=df,y='Year',order=order)
plt.title('Year of Graduation of Teachers',size=20);
```

```python
plt.figure(figsize=[20,30])
order= df['University Currently Teaching'].value_counts().index
sns.countplot(data=df,y='University Currently Teaching',order=order)
plt.title(' Teachers University Currently Teaching ',size=20);
```

```python
plt.pie(df['Province University Located'].value_counts(), labels=df['Province University Located'].value_counts().index,counterclock=False,autopct='%.2f')
plt.title('Teachers BY Province University Located');
```
