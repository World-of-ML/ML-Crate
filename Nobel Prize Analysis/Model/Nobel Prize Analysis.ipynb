# Data Visualization

```python
plt.hist(df['age_get_prize']);
plt.xlim(0, 100);
plt.xlabel('age when they get Nobel Prize');
plt.ylabel('awards Count')
plt.title("ages distribution", fontsize=22);
```

```python
plt.pie(df['gender'].value_counts(), labels=df['gender'].value_counts().index,counterclock=False,autopct='%.2f')
plt.title('awards BY GENDER');
```

```python
base_color = sns.color_palette()[0]
order= df['category'].value_counts().index
sns.countplot(data=df,y='category',order=order, color=base_color)
plt.title('awards BY category',size=20);
```

```python
plt.figure(figsize=[20,30])
order= df['born_country_code'].value_counts().index
sns.countplot(data=df,y='born_country_code',order=order)
plt.title('awards BY COUNTRY',size=20);
```

```python
plt.figure(figsize=[10,5])
sns.countplot(data=df,x='decade',hue='gender')
plt.title('Number of men and women won Nobel by decade',size=20);
```

```python
order= df['category'].value_counts().index
sns.countplot(data=df,y='category',order=order,hue='gender')
plt.title('Number of men and women won Nobel by category',size=20);
```

```python
order= df['born_country_code'].value_counts().index
plt.figure(figsize=[20,40])
sns.countplot(data=df,y='born_country_code',order=order,hue='gender')
plt.title('number of men and women won Nobel by Country',size=20);
```
