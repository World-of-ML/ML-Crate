```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
```

```python
fig, ax = plt.subplots(1,1, figsize=(12, 7), dpi=72)
sns.regplot(data=df, x='Length', y='Width', ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots(1,1, figsize=(12, 7), dpi=72)
sns.regplot(data=df, x='Length', y='Height', ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots(1,1, figsize=(12, 7), dpi=72)
sns.regplot(data=df, x='Width', y='Height', ax=ax)
plt.show()
```

```python
plt.figure(figsize=(35,10))
df['Height'].value_counts().plot.bar(title='Height',color='lightblue')
```

```python
plt.figure(figsize=(30, 20))
df.groupby('Operator')['Height'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(30, 20))
df.groupby('Operator')['Width'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(30, 20))
df.groupby('Operator')['Length'].mean().sort_values(ascending=False).plot(kind='bar')
```

```python
plt.figure(figsize=(20, 20))
top = df['Operator'].value_counts().sort_values(ascending=False)
top.plot(kind='pie')
```

```python
plt.figure(figsize=(50, 50))
top = df['Height'].value_counts().sort_values(ascending=False)
top.plot(kind='pie')
```
