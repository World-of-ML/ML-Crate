## Feature Engineering

We will add some additional feature for some of the dataframes that could be useful in our analysis and modeling

### Points_x / Points_y
total points so far for the home and away teams

```python
def add_points_feature(df):
    df['Points_x'] = 0
    df['Points_y'] = 0
        
    for index , row in df.sort_values(by='Date').iterrows():
        if row['Score_x'] > row['Score_y']:
            df.at[index , 'Points_x'] += 3
        elif row['Score_x'] < row['Score_y']:
            df.at[index , 'Points_y'] += 3
        else:
            df.at[index , 'Points_x'] += 1
            df.at[index , 'Points_y'] += 1
```

```python
add_points_feature(bundesliga_23_24)
add_points_feature(bundesliga_22_23)
add_points_feature(bundesliga_21_22)
```

### Goals_for_x / Goals_for_y

Total goals scored by home and away teams during the season

```python
def add_goals_for_feature(df):
    df['Goals_for_x'] = 0
    df['Goals_for_y'] = 0

    for index , row in df.sort_values(by = 'Date').iterrows():
        df.at[index , 'Goals_for_x'] += row['Score_x']
        df.at[index , 'Goals_for_y'] += row['Score_y']
```

```python
add_goals_for_feature(bundesliga_23_24)
add_goals_for_feature(bundesliga_22_23)
add_goals_for_feature(bundesliga_21_22)
```

### Goals_against_x / Goals_against_y

Total goals conceeded by home and away teams during the season

```python
def add_goals_against_feature(df):
    df['Goals_against_x'] = 0
    df['Goals_against_y'] = 0

    for index , row in df.sort_values(by = 'Date').iterrows():
        df.at[index , 'Goals_against_x'] += row['Score_y']
        df.at[index , 'Goals_against_y'] += row['Score_x']
```

```python
add_goals_against_feature(bundesliga_23_24)
add_goals_against_feature(bundesliga_22_23)
add_goals_against_feature(bundesliga_21_22)
```

### Goal Difference

```python
bundesliga_23_24['Goal_diff_x'] = bundesliga_23_24['Goals_for_x'] - bundesliga_23_24['Goals_against_x']
bundesliga_22_23['Goal_diff_x'] = bundesliga_22_23['Goals_for_x'] - bundesliga_22_23['Goals_against_x']
bundesliga_21_22['Goal_diff_x'] = bundesliga_21_22['Goals_for_x'] - bundesliga_21_22['Goals_against_x']
```

```python
bundesliga_23_24['Goal_diff_y'] = bundesliga_23_24['Goals_for_y'] - bundesliga_23_24['Goals_against_y']
bundesliga_22_23['Goal_diff_y'] = bundesliga_22_23['Goals_for_y'] - bundesliga_22_23['Goals_against_y']
bundesliga_21_22['Goal_diff_y'] = bundesliga_21_22['Goals_for_y'] - bundesliga_21_22['Goals_against_y']
```

We can use the engineered features to show the league standings

```python
def show_league_standings(df):
    df_points = (df.groupby('Home')['Points_x'].sum() + df.groupby('Away')['Points_y'].sum()).reset_index(name = 'Total Points')
    df_gf = (df.groupby('Home')['Goals_for_x'].sum() + df.groupby('Away')['Goals_for_y'].sum()).reset_index(name = 'Goals for')
    total_df = pd.merge(df_points , df_gf , on = 'Home' , how = 'inner')
    df_ga = (df.groupby('Home')['Goals_against_x'].sum() + df.groupby('Away')['Goals_against_y'].sum()).reset_index(name = 'Goals against')
    total_df = pd.merge(total_df , df_ga , on = 'Home' , how = 'inner')
    df_gd = df_ga = (df.groupby('Home')['Goal_diff_x'].sum() + df.groupby('Away')['Goal_diff_y'].sum()).reset_index(name = 'Goals Difference')
    total_df = pd.merge(total_df , df_gd , on = 'Home' , how = 'inner')
    return total_df.sort_values(by = 'Total Points' , ascending = False)
```

2022-2023 standings

### Result 
the result of the match, has one of 3 values: `win_x` , `win_y` , `draw`

```python
def get_match_results(row):
    if row['Score_x'] > row['Score_y']:
        return 'Win_x'
    elif row['Score_x'] < row['Score_y']:
        return 'Win_y'
    else:
        return 'Draw'

teams_df['result'] = teams_df.apply(get_match_results, axis=1)
```

## Predicting Match Results

We will build a Machine learning model that predicts match results using other features 

Our target will be the `result` column, we will make a numeric version of it to use in our analysis

```python
result_map = {'Win_x': 0 , 'Draw': 1 , 'Win_y': 2}
teams_df['result_numeric'] = teams_df['result'].map(result_map)
```

We will drop the features that are either irrelevant like game_id and Date or can be used to know match results directly like score_x and score_y

```python
teams_df_pred = teams_df.drop(['Day', 'Date', 'Time' , 'game_id' , 'Goal Kicks_y' , 'Throw Ins_y' , 'Long Balls_y' , 
                  'Goal Kicks_x' , 'Throw Ins_x' , 'Long Balls_x' , 'Score_x' , 'Score_y','Points_x' , 'Points_y' , 'result'] , 
                  axis = 1)
```

###  Encoding Categorical Features

```python
teams_df_pred =  pd.get_dummies(teams_df_pred, columns=teams_df_pred.select_dtypes('object').columns).astype(float)
teams_df_pred.head()
```

## Train test split

```python
X_train = pd.concat([teams_df_pred[teams_df_pred['season_2021-2022'] == 1] , teams_df_pred[teams_df_pred['season_2022-2023'] == 1][:153]])
X_val = teams_df_pred[teams_df_pred['season_2022-2023'] == 1][153:]
X_test = teams_df_pred[teams_df_pred['season_2023-2024'] == 1]
```

```python
X_train.shape , X_val.shape , X_test.shape
```

```python
y_train = pd.concat([teams_df[teams_df['season'] == '2021-2022'] , teams_df[teams_df['season'] == '2022-2023'][:153]])['result']
```

```python
y_val = teams_df[teams_df['season'] == '2022-2023'][153:]['result']
```

```python
y_test = teams_df[teams_df['season'] == '2023-2024']['result']
```

## Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.fit_transform(X_val)
X_test_scaled = scaler.fit_transform(X_test)
```

## Training a logistic regression model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
model = LogisticRegression(multi_class='ovr', solver='liblinear')

# Train the model
model.fit(X_train_scaled, y_train)
```

```python
predictions = model.predict(X_train_scaled)

# Evaluate the model
accuracy = accuracy_score(y_train, predictions)
print(f'Training Accuracy: {accuracy:.2f}')

# Additional evaluation metrics
print(classification_report(y_train, predictions))
```

## Evaluating on validation data

```python
# Make predictions on the val set
predictions = model.predict(X_val_scaled)

# Evaluate the model
accuracy = accuracy_score(y_val, predictions)
print(f'Validation Accuracy: {accuracy:.2f}')

# Additional evaluation metrics
print(classification_report(y_val, predictions))
```

```python
sns.heatmap(confusion_matrix(y_val, predictions) , annot=True, fmt='d', cmap='Blues' , 
            xticklabels=['Win_x', 'Draw', 'Win_y'],
            yticklabels=['Win_x', 'Draw', 'Win_y'])
plt.title("Validation confusion matrix for logistic regression model")
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

The model performs a bit poorly in predicting a win for the home team

## Evaluating on the test set

```python
# Make predictions on the val set
predictions = model.predict(X_test_scaled)

# Evaluate the model
accuracy = accuracy_score(y_test, predictions)
print(f'Test Accuracy: {accuracy:.2f}')

# Additional evaluation metrics
print(classification_report(y_test, predictions))
```

```python
sns.heatmap(confusion_matrix(y_test, predictions) , annot=True, fmt='d', cmap='Blues' , 
            xticklabels=['Win_x', 'Draw', 'Win_y'],
            yticklabels=['Win_x', 'Draw', 'Win_y'])
plt.title("Validation confusion matrix for logistic regression model")
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
