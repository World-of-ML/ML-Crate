# Visual Representation of accuracy scores of different machine learning models that are being applied for medical charges prediction

### Importing the required libraries


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

##  LINE PLOT

```python
# Accuracy scores
accuracy_scores = [0.877973, 0.868598, 0.864436]
model_names = ['Gradient Boosting Regressor','CatBoosting Regressor', 'Random Forest Regressor']

# Set the figure size
plt.figure(figsize=(8, 6))

# Create the line plot
plt.plot(model_names, accuracy_scores, marker='o', linestyle='-', color='b', linewidth=2)

# Add data labels with adjusted vertical alignment and offset
offset = 0.0001 
for i in range(len(model_names)):
    plt.text(model_names[i], accuracy_scores[i]+offset, str(accuracy_scores[i]), ha='center')

# Add labels and title
plt.xlabel('Model')
plt.ylabel('Accuracy Score')
plt.title('Accuracy Scores of Top 3 Models')

# Show the plot
plt.show()
```

## DISTRIBUTION PLOT

```python
# Set the figure size
plt.figure(figsize=(8, 6))

# Plot a histogram of the scores
plt.hist(accuracy_scores, bins=10, color='b', alpha=0.7)
plt.xlabel('Accuracy Score')
plt.ylabel('Frequency')

# Set title and show plot
plt.title('Distribution of Accuracy Scores')
plt.show()
```

## HEATMAP 

```python
data = np.array(accuracy_scores).reshape(1, -1)

# Set the figure size
plt.figure(figsize=(8, 6))

# Create the heatmap
sns.heatmap(data, cmap='Blues', annot=True, xticklabels=model_names, yticklabels=False, cbar=False)
plt.xlabel('Model')
plt.ylabel('')

# Set title and show plot
plt.title('Accuracy Scores of Different Models')
plt.show()
```
