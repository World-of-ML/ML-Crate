```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Eurovision Winners - Winner Data KG-3.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
#Checking for missing values
df.isnull().sum()
```

```python
#dropping the 'semi final points' column
df = df.drop(labels = 'Semi Final Points', axis = 1)
```

```python
#dropping the 'location' column also as 'country' is already mentioned
df = df.drop(labels = 'Location', axis = 1)
```

```python
#removing the row where 'total Points' is null
df.dropna(subset =['Total Points'], inplace = True)
```

```python
df.isnull().sum()
```

```python
#getting the statistical measures of the dataset
df.describe()
```

```python
#visualizing total points distribution
plt.figure(figsize = (6,6))
sns.displot(df['Total Points'])
plt.title("Total Points Distribution")
plt.savefig("Total Points Distribution.png")
plt.show()
```

```python
#kernel density estimate of the 'total points'
plt.figure(figsize = (6,6))
sns.kdeplot(df['Total Points'],linestyle = 'dotted')
plt.title("Kernel Density Estimate of the Total Points")
plt.savefig("Kernel Density Estimate of the Total Points.png")
plt.show()
```

```python
#comparing the total points scored against the year in which it was scored
plt.bar(df['Year'], df['Total Points'], align = 'center', color = 'purple')
plt.title("Total Points scored over the years")
plt.xlabel("Years")
plt.ylabel("Total Points")
plt.savefig("Total Points scored over the years.png")
plt.show()
```

```python
#visualizing the country column
plt.figure(figsize = (5,5))
sns.countplot(x = 'Country', data = df)
plt.xticks(rotation=90)
plt.title("Number of contests held in different countries")
plt.savefig("Number of contests held in different countries.png")
plt.show()
```

```python
#plotting the year of contest with the country it was held in
plt.scatter(df['Country'], df['Year'], marker ="*" )
plt.xlabel("Country")
plt.xticks(rotation=90)
plt.ylabel("Year")
plt.grid(True)
plt.title("year of contest and country it was held in")
plt.savefig("year of contest and country it was held in.png")
plt.show()
```

```python
#visualizing the 'running order' column
plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True
f, axes = plt.subplots(1, 2)

sns.histplot(df['Running Order'], ax=axes[0])
sns.kdeplot(df['Running Order'],linestyle = 'dotted', ax=axes[1])  #kde--->kernel density estimate
plt.savefig("Histplot and KDE of 'Running Order'.png")
plt.show()
```

```python
df1 = df[['Country','Total Points']].copy()
df1 = df1.sort_values("Total Points")

plt.bar(df1['Country'], df1['Total Points'], align = 'center')
plt.title("Countries sorted by total points scored in the Eurovision Contests held in them")
plt.xlabel("Countries")
plt.ylabel("Total Points")
plt.xticks(rotation=90)
plt.savefig("Countries sorted by total points scored in the Eurovision Contests held in them'.png")
plt.show()
```

```python
df2 = df[['Song','Total Points']].copy()
df2 = df2.sort_values("Total Points")

plt.figure(figsize = (15,8))
plt.bar(df2['Song'], df2['Total Points'], align = 'center', color = 'orange')
plt.title("Songs vs the points they scored")
plt.xlabel("Songs")
plt.ylabel("Total Points")
plt.xticks(rotation=90)
plt.savefig("Songs vs the points they scored.png")
plt.show()
```

```python
df3 = df[['Songwriter(s)','Total Points']].copy()
df3 = df3.sort_values("Total Points")

plt.figure(figsize = (25,8))
plt.bar(df3['Songwriter(s)'], df3['Total Points'], align = 'center', color='green')
plt.title("Songwriter(s) vs the points the scored by their songs")
plt.xlabel("Song writers")
plt.ylabel("Total Points")
plt.xticks(rotation=90)
plt.savefig("Songwriter(s) vs the points the scored by their songs.png")
plt.show()
```

```python
df4 = df[['Artist','Total Points']].copy()
df4 = df4.sort_values("Total Points")

plt.figure(figsize = (15,8))
plt.bar(df4['Artist'], df4['Total Points'], align = 'center', color ='gray')
plt.title("Artists v/s the points scored by them")
plt.xlabel("Artists")
plt.ylabel("Total Points")
plt.xticks(rotation=90)
plt.savefig("Artists vs the points scored by them.png")
plt.show()
```
